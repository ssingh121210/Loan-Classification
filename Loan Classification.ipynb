```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('credit_train.csv')
del df['Loan ID']
del df['Customer ID']
```

```python
# df.info()
df=df[0:100000]
```

```python
df.columns[df.isnull().any()]
```

```python
l=[1,4,5,6]
```

```python
for col in ['Loan Status','Term','Years in current job', 'Home Ownership','Purpose',]:
    df[col] = df[col].fillna('None')
```

```python
for cil in ['Current Loan Amount','Credit Score','Annual Income','Monthly Debt', 'Years of Credit History',
       'Months since last delinquent', 'Number of Open Accounts',
       'Number of Credit Problems', 'Current Credit Balance',
       'Maximum Open Credit', 'Bankruptcies', 'Tax Liens']:
    df[cil]=df[cil].fillna(int(0))
```

```python
df
```

```python
y=df.iloc[:,0].values
y
```

```python
X=df.iloc[:,1:].values
X
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label_x=LabelEncoder()
for m in l:   
    X[:,m]=label_x.fit_transform(X[:,m])
    onehotencoder=OneHotEncoder(categorical_features=[m])
```

```python
X[0]
```

```python
X=onehotencoder.fit_transform(X).toarray()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
label_x=LabelEncoder()
y=label_x.fit_transform(y)
onehotencoder=OneHotEncoder(categorical_features=[])
```

```python
#to show full array
import sys
import numpy
numpy.set_printoptions(threshold=sys.maxsize)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30,random_state=42)
```

# Classifiers

# 1 Logistic Regression

```python
from sklearn.preprocessing import StandardScaler
sc_X=StandardScaler()
X_test=sc_X.fit_transform(X_test)
X_train=sc_X.fit_transform(X_train)
```

```python
from sklearn.linear_model import LogisticRegression
classifier=LogisticRegression(random_state=42)
```

```python
classifier.fit(X_train,y_train)
```

```python
y_pred=classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
```

```python
cm
```

```python
from sklearn.metrics import classification_report
```

```python
target_names = ['class 0', 'class 1']
print(classification_report(y_test, y_pred,target_names=target_names))
```


## Naive Bias

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classifierNB=GaussianNB()
classifierNB.fit(X_train,y_train)
```

```python
y_predNB=classifierNB.predict(X_test)
```

```python
cm
```

## DECISION TREE

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifierDT=DecisionTreeClassifier(criterion='entropy',random_state=42)
classifierDT.fit(X_train,y_train)
```

```python
y_predDT=classifierDT.predict(X_test)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
classifierRF=RandomForestClassifier(n_estimators=10, random_state=0,criterion='entropy')
classifierRF.fit(X_train,y_train)
```

```python
y_predRF=classifierRF.predict(X_test)
```
